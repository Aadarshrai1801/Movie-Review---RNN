### Implementation of Word Embedding

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing.text import one_hot #type: ignore
from tensorflow.keras.layers import Embedding #type: ignore
from tensorflow.keras.preprocessing.sequence import pad_sequences #type: ignore
from tensorflow.keras.models import Sequential #type: ignore
```

```python
# Sentences
sentence = [
    "The glass of milk",
    "The glass of juice",
    "I am a good boy",
    "I am a good developer",
    "Understand the meaning of words",
    "Your videos are good"
]
```

```python
# Define the vocabulary size
voc_size = 10000
```

```python
# One Hot  representation
onehot_rep = [one_hot(words, voc_size) for words in sentence]
```

```python
onehot_rep
```

```python
# Word embedding representation
sentence_length = 8
```

```python
embedded_docs = pad_sequences(onehot_rep, padding="pre", maxlen=sentence_length)
```

```python
print(embedded_docs)
```

```python
# Feature representation
dimension = 10
```

```python
model = Sequential()
model.add(Embedding(voc_size, dimension, input_length=sentence_length))
```

```python
model.compile("adam", "mse")
```

```python
model.summary()
```

```python
model.predict(embedded_docs[0])
```
